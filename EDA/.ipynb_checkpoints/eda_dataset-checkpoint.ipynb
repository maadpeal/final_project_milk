```python
import pandas as pd
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_json('../Resources/full.json')
```

```python
# how many cows are there?
print(f"There are: {len(df['id'].unique())} cows")
```

```python
# how many cows had a profitable milking cycle?
print(f"There are: {len(df.query('daily_production >= 10000'))} profitable milking cycle")
print(f"There are: {len(df.query('daily_production < 10000'))} unprofitable milking cycle")
```

```python
objects = ('Profitable', 'Unprofitable')
y_pos = np.arange(len(objects))
performance = [len(df.query('daily_production >= 10000')),len(df.query('daily_production < 10000'))]

plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Count of milking cycle')
plt.title('Profitability Comparison')

plt.show()
```

```python
bins = list(range(df.daily_production.min(), df.daily_production.max()+250, 3000))
labels = [f'column{i}' for i in range(len(bins)-1)]
df['bins'] = pd.cut(df.daily_production, bins, labels=labels, include_lowest=True)
df['bins'].value_counts()
```

```python
values = list(df['bins'].value_counts().values)
objects = [
    '0_3000', 
    '9001_12000', 
    '6001_9000', 
    '12001_15000',
    '3001_6000',
    '15002_18000',
    '18004_20997',
    '21004_23984',
    '24032_26903',
    '27014_29844',
    '30029_32516',
    '35649',
    '37047'
]
```

```python
y_pos = np.arange(len(objects))
performance = values

plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects, rotation=60)
plt.ylabel('Counts of bins')
plt.title('Bins of Total production by lactation')

plt.show()
```
