```python
#estado ginecologico y media de dias de leche--opcional
gyn_status = mastitis_df.groupby(["gynecological_status"]).mean()["milk_days"]
gyn_status.head(10)
```

```python
#vacas por estado ginecologico 
cow_by_gyn_state = mastitis_df.groupby(["gynecological_status"]).count()["cow"]
cow_by_gyn_state.head(10)
```

```python
#grafico de pay para vacas por estado ginecologico 
plt.pie(cow_by_gyn_state,
        labels=["abortion", "heat", "insemination", "labor", "no_inseminate", "pregnacy"],
        colors=["blue", "lightskyblue", "darkblue", "navy", "steelblue", "mediumblue"],
        explode=[0,0,0,0,0.1,0.1],
        autopct='%1.1f%%',
        shadow=True, startangle=150)
plt.title("Numnero de vacas por estado ginecologico ")
plt.show()
```

```python
#numero de casos de mastitis en la primera lactancia por estado ginecologico
gyn_status_ill_count = mastitis_df.groupby(["gynecological_status"]).count()["number_mastitis_lactation_1"]
gyn_status_ill_count.head(10)
```

```python
#grafico de pay para mastitis en la primera lactancia por estado ginecologico
plt.pie(gyn_status_ill_count,
        labels=["abortion", "heat", "insemination", "labor", "no_inseminate", "pregnacy"],
        colors=["blue", "lightskyblue", "darkblue", "navy", "steelblue", "mediumblue"],
        explode=[0,0,0,0,0.1,0.1],
        autopct='%1.1f%%',
        shadow=True, startangle=150)
plt.title("Mastitis en la primera lactancia por estado ginecologico")
plt.show()
```

```python
#grafico numero de casos de mastitis en la primera lactancia por estado ginecologico
plt.scatter(gyn_status_ill_count, gyn_status)
plt.show()
```
