```python
import pandas as pd
```

```python
df = pd.read_csv('./Resources/Step_1.csv')
```

```python
# eliminar Cow, ID_Cow_Bracelet_Podometer
df_clean = df[['Patio', 'Gynecological_Status',
       'Birth_Date', 'Daily_Production', 'Average_Daily_Production_Days',
       'Average_Daily_Production_Percentage', 'Lactation_Number', 'Milk_Days',
       'Days_After_Heat', 'Insemination', 'Days_After_Insemination',
       'Pregnacy_Days', 'Inseminator_Name', "Bull's Name",
       'Number_of_Abortions', 'Insemination_Date', 'Production_Peak',
       'Previous_Lactation_Production', 'Actual_Lactation_Production']]
```

```python
df_clean.loc[df_clean['Daily_Production'] < 28, 'Daily_Production_bool'] = 0
```

```python
df_clean.loc[df_clean['Daily_Production'] >= 28, 'Daily_Production_bool'] = 1
```

```python
df_clean
```

```python
df_clean['Birth_Date'] = pd.to_datetime(df_clean['Birth_Date'])
```

```python
df_clean_a = df_clean[['Gynecological_Status', 'Birth_Date', 'Daily_Production',
       'Average_Daily_Production_Days', 'Average_Daily_Production_Percentage',
       'Lactation_Number', 'Milk_Days', 'Days_After_Heat', 'Insemination',
       'Days_After_Insemination', 'Pregnacy_Days', 'Inseminator_Name',
       "Bull's Name", 'Number_of_Abortions', 'Insemination_Date',
       'Production_Peak', 'Previous_Lactation_Production',
       'Actual_Lactation_Production', 'Daily_Production_bool']]
```

```python
df_clean_a = df_clean_a.rename(columns={'Birth_Date':'given_birthdate'})
df_clean_a
```

```python
from datetime import timedelta, datetime
```

```python
df_clean_a['given_birthdate_days'] = df_clean_a['given_birthdate'] - datetime(2022,4,6)
```

```python
df_clean_a.info()
```

```python
df_clean_a['given_birthdate_days'] = df_clean_a['given_birthdate_days'].dt.days
```

```python
df_clean_a['given_birthdate_days'] = df_clean_a['given_birthdate_days'] * -1
```

```python
df_clean_a = df_clean_a[['Gynecological_Status', 'Daily_Production',
       'Average_Daily_Production_Days', 'Average_Daily_Production_Percentage',
       'Lactation_Number', 'Milk_Days', 'Days_After_Heat', 'Insemination',
       'Days_After_Insemination', 'Pregnacy_Days', 'Inseminator_Name',
       "Bull's Name", 'Number_of_Abortions', 'Insemination_Date',
       'Production_Peak', 'Previous_Lactation_Production',
       'Actual_Lactation_Production', 'Daily_Production_bool',
       'given_birthdate_days']]
```

```python
#df_clean.loc[df_clean['Daily_Production'] >= 28, 'Daily_Production_bool'] = 1

df_clean_a.loc[df_clean_a['Inseminator_Name'] == '0', 'Inseminator_Name'] = 'N/A'
```

```python
df_clean_a.loc[df_clean_a["Bull's Name"] == '0', "Bull's Name"] = 'N/A'
```

```python
df_clean_a = df_clean_a[['Gynecological_Status', 'Daily_Production',
       'Average_Daily_Production_Days', 'Average_Daily_Production_Percentage',
       'Lactation_Number', 'Milk_Days', 'Days_After_Heat', 'Insemination',
       'Days_After_Insemination', 'Pregnacy_Days', 'Inseminator_Name',
       "Bull's Name", 'Number_of_Abortions',
       'Production_Peak', 'Previous_Lactation_Production',
       'Actual_Lactation_Production', 'Daily_Production_bool',
       'given_birthdate_days']]
```

```python
df_clean_a = df_clean_a.rename(columns={"Bull's Name":'bulls_name'})
```

```python
# Create the OneHotEncoder instance
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(sparse=False)
# Gynecological_Status, Inseminator_Name, bulls_name, 
df_clean_a
```

```python
# Fit the encoder and produce encoded DataFrame
encode_df = pd.DataFrame(enc.fit_transform(df_clean_a.Gynecological_Status.values.reshape(-1,1)))

# Rename encoded columns
encode_df.columns = enc.get_feature_names(['Gynecological_Status'])
encode_df.head()
```

```python
df_clean_a = df_clean_a.merge(encode_df,left_index=True,right_index=True).drop("Gynecological_Status",1)
df_clean_a
```

```python
# Fit the encoder and produce encoded DataFrame
encode_df = pd.DataFrame(enc.fit_transform(df_clean_a.Inseminator_Name.values.reshape(-1,1)))

# Rename encoded columns
encode_df.columns = enc.get_feature_names(['Inseminator_Name'])
encode_df.head()
```

```python
df_clean_a = df_clean_a.merge(encode_df,left_index=True,right_index=True).drop("Inseminator_Name",1)
df_clean_a
```

```python
# Fit the encoder and produce encoded DataFrame
encode_df = pd.DataFrame(enc.fit_transform(df_clean_a.bulls_name.values.reshape(-1,1)))

# Rename encoded columns
encode_df.columns = enc.get_feature_names(['bulls_name'])
encode_df.head()
```

```python
df_clean_a = df_clean_a.merge(encode_df,left_index=True,right_index=True).drop("bulls_name",1)
df_clean_a
```

```python
df_clean_a.to_csv('./Resources/data_prepared.csv')
```
