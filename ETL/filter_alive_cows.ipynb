```python
df = df.reset_index(drop=True)
df.fillna("nan")
```

```python
for i in range(1,10):
    df['lactation_' + str(i)] = 0
```

```python
for id in list_ids:
    filter_df = df.query(f'id == {id}')
    if filter_df.lactation.unique().any():
        for lact in list(filter_df.lactation.unique()):
            index_ = df.query(f'id == {id}').index[0]
            df.iloc[index_]['lactation_' + str(lact)] = 1
```

```python
for id in list_ids:
    filter_df = df.query(f'id == {id}')
    if filter_df.lactation.unique().any():
        for lact in list(filter_df.lactation.unique()):
            index_ = df.query(f'id == {id}').index[0]
            index_t = df.query(f'id == {id}').index[lact - 1]
            df.iloc[index_]['total_production_' + str(lact)] = df.iloc[index_t]['daily_production']
```

```python
for id in list_ids:
    filter_df = df.query(f'id == {id}')
    if filter_df.lactation.unique().any():
        for lact in list(filter_df.lactation.unique()):
            index_ = df.query(f'id == {id}').index[0]
            index_t = df.query(f'id == {id}').index[lact - 1]
            df.iloc[index_]['milk_days_' + str(lact)] = df.iloc[index_t]['milk_days']
```

```python
group = df.groupby('id').first()
```

```python
group_df = group.reset_index()
```

```python
group_df = group_df.drop(['lactation', 'daily_production', 'milk_days'], axis=1)
```

## merge

```python
mastitis_filter_to_merge = mastitis_filter[['vaca', 'mastitis_lac_1', 'mastitis_lac_2', 'mastitis_lac_3',
       'mastitis_lac_4', 'mastitis_lac_5', 'mastitis_lac_6', 'mastitis_lac_7',
       'mastitis_lac_8', 'mastitis_lac_9']]
inseminaciones_filter_to_merge = inseminaciones_filter[['vaca','inseminacion_lac_1', 'inseminacion_lac_2', 'inseminacion_lac_3',
       'inseminacion_lac_4', 'inseminacion_lac_5', 'inseminacion_lac_6',
       'inseminacion_lac_7', 'inseminacion_lac_8', 'inseminacion_lac_9']]
abortos_filter_to_merge = abortos_filter[['vaca', 'abortos_Lac_1', 'abortos_lac_2', 'abortos_lac_3', 'abortos_lac_4',
       'abortos_lac_5', 'abortos_lac_6', 'abortos_lac_7', 'abortos_lac_8',
       'abortos_lac_9']]

```

```python
total_cows_data = pd.merge(group_df, mastitis_filter_to_merge, how='left', left_on='id', right_on='vaca').fillna(False, downcast='infer')
```

```python
total_cows_data = pd.merge(total_cows_data, inseminaciones_filter_to_merge, how='left', left_on='id', right_on='vaca').fillna(False, downcast='infer')
```

```python
total_cows_data = pd.merge(total_cows_data, abortos_filter_to_merge, how='left', left_on='id', right_on='vaca').fillna(False, downcast='infer')
```

```python
total_cows_data = total_cows_data.drop(['vaca'], axis=1)
```
